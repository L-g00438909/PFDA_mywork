# formatting and parsing


the default string repreentation of a datetime object uses the format: 
-   YYYY-MM-DDTHH:MM:SS.mmmmmm
-   Alternative formats can be generated using ```strftime()``` fnction.
      -   converts datetime objects into strings
      -   uses different code control code to give an output.
      -   each control code resembles different parameters like year, month, weekday and date. 
-    if your input data includes timestamp values parsable with ```time.strptime()```, then ```datetime.strptime() is a convenient way to convert them to datetime instances. 

## Formatting: 

Weekday Month Day Hour:Minute:Second Year


````python
format = '%a %d %b %H:%M:%S %Y'

import time
import datetime
today= datetime.datetime.today()
print(f'ISO: {today}')

today_str = today.strftime(format)
print(f'Strftime: {today_str}')
````

````python
# time in HH:MM:SS

time_str = today.strftime('%H:%M:%S')
print(f'Time: {time_str}')

time_str = today.strftime('%X')
print(f'Time: {time_str}')

print(today.strftime('%p'))
````

### parsing:

if the format of a string is known it can be easily parsed to a datetime object using strptime

````python
datetime = datetime.datetime.strptime(today_str, format)
print(datetime)
````

````python
import datetime

datetime_str = 'Sep 19 2020 2:42PM'
datetime_fmt = '%b %d %Y %I:%M%p'
datetime = datetime.datetime.strptime(datetime_str, datetime_fmt)

print('Date', datetime.date())
print('Time', datetime.time())
print('Date-time', datetime)

# read in a string time and convert it to a datetime object
# parsing means to read in a string and convert it to a datetime object 
# strptime() function is used to parse a string representing a time according to a format
````
