Random Number Generation: is the process by which a sequence of pseudo random numbers may be drawn. 
1. they are drawn from a probability distribution. the most common one is the uniform distribution on the domaain.ie: random numbers between xero and one. completely random does not make sense because of the infinite magnitude of numbers. 

2. the random sequence depends on an input seed and are then generated by a deterministic algorithm from that seed

Numpy documentation: random generator. 

```python
#np.random is a module in numpy that can generate random numbers from many kinds of probability distributions
np.random.uniform(low=0, high=1, size=10)
```

The function uniform() in the np.random module generates random numbers on the interval[low, high] from a uniform distribution. the size parameter is how many random numbers and is all of Numpys random number generation. the function returns a NumPy array. 


```python
# generate random numbers
x = np.random.uniform(low=0, high=1, size=100
                      )
plt.hist(x)
```

```python
# to simulate flip of an unfair coin
size = 20 # number of flips
x = np.random.uniform(low=0, high=1, size=size) # generate random numbers between 0 and 1 for each flips
heads = x < 0.7 # if the random number is less than 0.7, it is heads. 70% chance of heads
print(heads)
heads_count=heads.sum() # count the number of heads
tails_count = size - heads_count # count the number of tails
print(f"Number of heads: {heads_count}")
print(f"Number of tails: {tails_count}")

labels =[ 'Heads', 'Tails'] 
plt.pie(np.array([heads_count, tails_count]), labels=labels, autopct='%1.1f%%')
```

#### Drawing random numbers out of other distributions

We can also draw random numbers from other probability distributions besides uniform. for example: distribution with mean u and standard deviation

```python
# set parameters
mu = 10 # mean
sigma =3 # standard deviation

# draw 100000 random numbers from a normal distribution
x = rng.normal(mu, sigma, 100000) # normal distribution with mean mu and standard deviation sigma and 100000 samples

# plot histogram
plt.hist(x, bins=1000,)
plt.show()
```

looks normal but again comparing the resulting ECDD is a better way to look at this. We'll check out the ECDF with 1000 samples so as not to choke the browser with the display.

```python
# choosing elements from an array randomly
# the rng.choice() function can be used to choose elements from an array randomly
rng= np.random.default_rng(seed=126969234)
rng.integers(0, 10, size=10) # generate 10 random integers between 0 and 10
```

```python
rng.choice(np.arange(10), size=10, replace=False) # choose 10 elements from an array of 10 elements. replace = False means that the elements are not replaced  

# with choice replace will never get same number twice.
```

rng.choice() can use replace to specify that you cant draw something twice. if its taken out of the array its taken out of the array 

```python
# random DNA sequence

''.join(rng.choice(list('ACGT'), size=100, replace=True)) # generate a random DNA sequence of length 10
```
