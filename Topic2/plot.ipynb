```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# np.arrange will create a series of numbers from 0 to 10 with a step of 0.1
x = 0.5 + np.arange(8)
y = [1,3,5,8,3,6,4,6]
y2 = [3,5,9.4,9.1,4.6,7.7,8.1,9] # this is the data we want to plot
```

```python
# make a plot(basic with nothing in it.)

fig, ax = plt.subplots()

# make plot-bar
#ax.bar(x,y)
#ax.plot(x,y)
#ax.scatter(x,y)
#ax.fill_between(x,y,y2)
#ax.stackplot(x,y,y2)
#ax.stairs(y)
#ax.stem(x,y)

# can put two plots on at same time. bar plt with x and y and line plot with x and y2
ax.bar(x,y)
ax.plot(x,y2)
plt.show() # once you do show you can't add anything to the plot after that.
```

```python
#matplotlib documenation-axes for setting the axis

fig, ax = plt.subplots()

ax.bar(x,y)
ax.plot(x,y2)

ax.set(xlabel= 'x-axis', ylabel= 'y-axis', title= 'Title')
```

```python
# can also format the labels of the x  y axis. for example money format

fig, ax = plt.subplots()
def currencyformatter (x, pos):    #position is the position of the tick on the axis
    return f"${x}M    " # f-string to format the number to a string with a dollar sign and M at the end of it   
```

```python
fig, ax = plt.subplots()

ax.bar(x,y)
ax.plot(x,y2)

ax.set_title('Title')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')

ax.xaxis.set_major_formatter(currencyformatter)
plt.show()
```

```python
# code to put labels at an angle. get tick labels and rotate them by 45 degrees 
fig, ax = plt.subplots()

ax.bar(x,y)
ax.plot(x,y2)

ax.set_title('Title')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')

ax.xaxis.set_major_formatter(currencyformatter)
plt.xticks(rotation=45) # rotate the x-axis labels by 45 degrees
plt.show()

# lifecycle matplotlib documentation 
```

```python
# can add text to any location you like inside a plot. 
fig, ax = plt.subplots()

ax.bar(x,y)
ax.plot(x,y2)

ax.set_title('Title')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')

ax.text(1,8, 'Text', fontsize=12, color='red', verticalalignment='center') # add text to the plot at the location (1,8) with the text 'Text' in red and fontsize 12


ax.xaxis.set_major_formatter(currencyformatter)

labels = ax.get_xticklabels() # get the x-axis labels
plt.setp(labels, rotation=45, horizontalalignment='right') # set the rotation of the x-axis labels to 45 degrees

plt.show()
```

```python
# Annotations are similar to text but they are anchored to a point in the plot.

fig, ax = plt.subplots()

ax.bar(x,y)
ax.plot(x,y2)

ax.set_title('Title')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')

#x.text(1,8, 'Text', fontsize=12, color='red', verticalalignment='center') # add text to the plot at the location (1,8) with the text 'Text' in red and fontsize 12

ax.annotate('Annotation', (3.2,8), xytext=(1,9),arrowprops=dict(facecolor='black', shrink=0.05)) # add an annotation to the plot at the location (1,8) with the text 'Annotation' in red and fontsize 12

ax.xaxis.set_major_formatter(currencyformatter)

labels = ax.get_xticklabels() # get the x-axis labels
plt.setp(labels, rotation=45, horizontalalignment='right') # set the rotation of the x-axis labels to 45 degrees

plt.show()

# can add a legend to the plot.

fig, ax = plt.subplots()

# check matplotlib documentation for the different types of plots that can be made
```

```python
# matplot-anatomy.py 

# tick labels are the labels on the x and y axis. 
# tick marks are the lines on the x and y axis that show the location of the tick labels.
# spine is the border of the plot
# grid is the grid lines that can be added to the plot
# legend is the box that shows the labels of the different plots on the plot

# can change the color of the spine, grid, and tick marks by setting the color attribute of the spine, grid, and tick marks
# can change the width of the spine, grid, and tick marks by setting the linewidth attribute of the spine, grid, and tick marks
# can change the style of the spine, grid, and tick marks by setting the linestyle attribute of the spine, grid, and tick marks
# can change the color of the tick labels by setting the color attribute of the tick labels
# can change the font size of the tick labels by setting the fontsize attribute of the tick labels
# can change the font style of the tick labels by setting the fontstyle attribute of the tick labels
# can change the font weight of the tick labels by setting the fontweight attribute of the tick labels
```

```python
# function to create a circle with a given radius and center point 
def circle(radius, center):
    circle = plt.Circle(center, radius, color='black', fill=False)
    return circle

# function to create a rectangle with a given width, height, and center point
def rectangle(width, height, center):
    rectangle = plt.Rectangle((center[0] - width/2, center[1] - height/2), width, height, color='black', fill=False)
    return rectangle
# function to create a line with a given start and end point
def line(start, end):
    line = plt.Line2D((start[0], end[0]), (start[1], end[1]), color='black')
    return line

# create a plot with a circle, rectangle, and line
fig, ax = plt.subplots()

ax.add_artist(circle(1, (0, 0)))
ax.add_artist(rectangle(2, 3, (0, 0)))
ax.add_artist(line((-2, -2), (2, 2)))

# set the limits of the plot
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)

# set the aspect of the plot to be equal
ax.set_aspect('equal')

# set the grid to be visible
ax.grid(True)

# set the x and y axis to be visible
ax.spines['left'].set_visible(True)
ax.spines['bottom'].set_visible(True)

# set the color of the x and y axis to be black
ax.spines['left'].set_color('black')

# set the width of the x and y axis to be 2
ax.spines['left'].set_linewidth(2)

# set the style of the x and y axis to be dashed
ax.spines['left'].set_linestyle('dashed')

# set the color of the tick labels to be black
ax.tick_params(axis='x', colors='black')    # set the color of the x-axis tick labels to be black
ax.tick_params(axis='y', colors='black')    # set the color of the y-axis tick labels to be black

# set the font size of the tick labels to be 12
ax.tick_params(axis='x', labelsize=12)    # set the font size of the x-axis tick labels to be 12
ax.tick_params(axis='y', labelsize=12)    # set the font size of the y-axis tick labels to be 12





plt.show()

# look up anatomy of a figure in matplotlib documentation

```
