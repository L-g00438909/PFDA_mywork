# Subplots

```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
# we can create subplots by passing parameters to the subplot() function. 
# The subplot() function takes three parameters: nrows, ncols, index 
# where nrows and ncols are the number of rows and columns of the subplot grid and index is the index of the subplot.

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(5,5))
plt.show()
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(8,4))
axs[0,1].plot([1,2,3]),([3,2,1])
```

```python
# for loop to plot in all subplots
num_cols=3
num_rows=2
fig, axs = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(8,4))
for row in range(num_rows):
    for col in range(num_cols):
        axs[row,col].annotate(f'plot[{row},{col}]', (0.5, 0.5), ha='center')
plt.show()
```

```python
# example:

data = [
    [8,7,6,5,4,3,2,1],
    [1,2,3,4,5,6,7,8],   
    [5, 5, 5, 5, 5, 5, 5, 5],
    [2,4,5,6,3,5,6,7]        
]
x=0.5 +np.arange(8)



num_cols=2
num_rows=2
fig, axs = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(8,4))
count = 0
for row in range(num_rows):
    for col in range(num_cols):
        axs[row,col].plot(x, data[count])
        count += 1

plt.show()
```
